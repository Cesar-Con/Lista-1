```python
altura = float(input("Digite a altura (cm): "))
largura = float(input("Digite a largura (cm): "))
profundidade = float(input("Digite a profundidade (cm): "))
volume = altura * largura * profundidade
print (f"O volume da caixa é (volume) cm³!!!")
```
